### Which product category got the maximum revenue

```python
df["Category"] = df["Product_Category"] + "," +df["Sub_Category"]
```

```python
df.head()
```

```python
category = df.groupby("Category").mean()
category
```

```python
### Plotting the data
pro_cat = np.sort(df["Category"].unique())
plt.bar(pro_cat,category["Revenue"])
plt.xticks(pro_cat,rotation="vertical")
plt.title("Category V/S Revenue")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.show()
```

```python
df.head()
```

### What age group buy which type of products

```python
category["Customer_Age"] = category["Customer_Age"].astype("int32")
age_pro_cat = np.sort(df["Category"].unique())
plt.bar(pro_cat,category["Customer_Age"])
plt.xticks(pro_cat,rotation="vertical")
plt.title("Category V/S Customer_age")
plt.xlabel("Category")
plt.ylabel("Age")
plt.show()
```

###   What category of products were bought with maximum category

```python
category = df.groupby("Category").sum()
pro_cat = np.sort(df["Category"].unique())
plt.bar(pro_cat,category["Order_Quantity"])
plt.xticks(pro_cat,rotation="vertical")
plt.title("Category V/S Quantity")
plt.xlabel("Category")
plt.ylabel("Quantity")
plt.show()
```

```python
### Sales per year do we have
year = np.sort(df["Year"].unique())
sales = df.groupby("Year").count()
pro_sale = list(sales["Product"])
for i in range (0,len(year)):
    print(f"Sales in year {year[i]} : {pro_sale[i]}")
```
